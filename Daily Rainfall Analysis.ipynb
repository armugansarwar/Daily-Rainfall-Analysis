```python
import numpy as np                                       # Importing the libraries
import scipy as sp
from scipy import io
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn import tree
```

### Opening .mat file

```python
mat_file = sp.io.loadmat('2010rainfall.mat')             # Opening the training data
XR1 = (mat_file['XR1'])
ZR1 = (mat_file['ZR1'])
n1 = XR1.shape[0]                                        # Number of locations
n2 = XR1.shape[1]                                        # Number of days
```

### Training using 2010 Data

```python
def train_rain_data(s) :                                 # s is the city number. Hence city index will be (s-1)
    X = np.zeros(((n1-1)+2,n2-2))
    l = 0                                                # Variable to traverse through a column of X
    for k in range(n2) :                                 # Variable to traverse through a column of XR1
        if k==0 or k==1 : continue                       # Excluding first 2 days
        j = 0
        for i in range(n1) :
            if i==s-1 : continue                         # Excluding selected location
            X[j,l] = XR1[i,k-1]                          # Rainfall at all other locations on that day
            j = j+1
        X[-2,l] = XR1[s-1,k-3]                           # Rainfall at that location 2 days before
        X[-1,l] = XR1[s-1,k-2]                           # Rainfall at that location 1 day before
        l = l+1
    Y = np.zeros((1,n2-2))
    l = 0
    for k in range(n2) :
        if k==0 or k==1 : continue
        Y[0,l] = XR1[s-1,k]
        l = l+1
    return (X.T),(Y.T)

#X,y = rain_data(42)
#print(X.shape,y.shape)
```

```python
X,y = train_rain_data(42)                               # For Mumbai (s=42)
reg1 = LinearRegression().fit(X,y)             
X,y = train_rain_data(158)                              # For Delhi (s=158)
reg2 = LinearRegression().fit(X,y)             
X,y = train_rain_data(299)                              # For Kharagpur (s=299)
reg3 = LinearRegression().fit(X,y)         
```

### Testing on 2011 Rainfall Data

```python
mat_file = sp.io.loadmat('2011rainfall.mat')            # Opening the test file
XR = (mat_file['XR'])
ZR = (mat_file['ZR'])
n1 = XR.shape[0]                                        # Number of locations
n2 = XR.shape[1]                                        # Number of days
#print(n1,n2)
```

```python
def test_rain_data(s) :                                 # s is the city number. Hence city index will be (s-1)
    X = np.zeros(((n1-1)+2,n2-2))
    l = 0                                               # Variable to traverse through a column of X
    for k in range(n2) :                                # Variable to traverse through a column of XR
        if k==0 or k==1 : continue                      # Excluding first 2 days
        j = 0
        for i in range(n1) :
            if i==s-1 : continue                        # Excluding selected location
            X[j,l] = XR[i,k-1]                          # Rainfall at all other locations on that day
            j = j+1
        X[-2,l] = XR[s-1,k-3]                           # Rainfall at that location 2 days before
        X[-1,l] = XR[s-1,k-2]                           # Rainfall at that location 1 day before
        l = l+1
    Y = np.zeros((1,n2-2))
    l = 0
    for k in range(n2) :
        if k==0 or k==1 : continue
        Y[0,l] = XR[s-1,k]
        l = l+1
    return (X.T),(Y.T)               # Transpose is returned because scikit takes input in the shape(examples,features)
```
